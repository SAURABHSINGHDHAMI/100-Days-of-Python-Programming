```python
L = [11,14,27,21,23,56,78,39,45,29,28,30]

# sum of even numbers
# sum of odd numbers
# sum of no.s divisible by 3

def return_sum(L):
    
    even_sum = 0
    odd_sum = 0
    div3_sum = 0
    
    for i in L:
        if i%2 == 0:
            even_sum = even_sum + i
        
    for i in L:
        if i%2 != 0:
            odd_sum = odd_sum + i
        
    for i in L:
        if i%3 == 0:
            div3_sum = div3_sum + i
    
    return(even_sum, odd_sum, div3_sum)

print(return_sum(L))
```

```python
L = [11,14,27,21,23,56,78,39,45,29,28,30]

def return_sum(func, L):
    
    result = 0
    
    for i in L:
        
        if func(i):
            result = result + i
        
    return result

x = lambda x : x%2 == 0
y = lambda x : x%2 != 0
z = lambda x : x%3 == 0

print(return_sum(x, L))
print(return_sum(y, L))
print(return_sum(z, L))
```

### Higher-Order Functions

1. Map
2. Filter
3. Reduce

###### 1. Map

```python
# map() function takes in a function + list as an argument and returns a new list with lambda modified items returned by that function for each item.
```

```python
L = [1,2,3,4,5,6,7]
L
```

```python
map(lambda x : x * 2, L)
```

```python
list(map(lambda x : x * 2, L))
```

```python
list(map(lambda x : x % 2 == 0, L))
```

```python
students = [
    {
        "name" : "Jacob Martin",
        "Father name" : "Ros Martin",
        "Address" : "123 Hills Street",
    },{
        "name" : "Angela Stevens",
        "Father name" : "Robert Stevens",
        "Address" : "3 Upper Street London",
    },{
        "name" : "Ricky Smart",
        "Father name" : "William Smart",
        "Address" : "Unknown",
    }
    
]

list(map(lambda student : student["name"], students))
```

###### 2. Filter

```python
# filter() function takes in a function + list as arguments and filters out all the elements of a sequence “sequence”, for which the function returns True.
```

```python
L
```

```python
list(filter(lambda x : x > 4, L))
```

```python
fruits = ["Apple", "Orange", "Mango", "Guava"]
```

```python
list(filter(lambda fruit : "e" in  fruit, fruits))
```

###### List Comprehension

```python
# Python List comprehension provides a much more short syntax for creating a new list based on the values of an existing list.
```

```python
L
```

```python
L1 = [item * 2 for item in L]
L1
```

```python
L2 = [i**2 for i in range(10)]
L2
```

```python
L3 = [i**2 for i in range(10) if i%2!=0]
L3
```

```python
fruits
```

```python
L4 = [fruit for fruit in fruits if fruit[0]=="O"]
L4
```

###### Dictionary Comprehension

```python
# same as list comprehension, dictionary comprehension creates dictionaries using simple expressions. 
# A dictionary comprehension takes the form {key: value for (key, value) in iterable}
```

```python
D = {"Name":"Nitish","Gender":"Male","Age":30}
```

```python
D.items()
```

```python
D1 = {key:value for key, value in D.items() if len(key)>3}
D1
```
