# Operators

Operators are used to perform operations on variables and values. Python has the following operators:
- Arithmetic operators
- Comparison operators/ Relational operator
- Logical operators
- Bitwise operators
- Assignment operators
- Identify operators
- Membership operators

## Arithmetic operators

```python
x = 5
y = 2

print(x + y)
```

```python
print(x - y)
```

```python
print(x * y)
```

```python
print(x / y)
```

```python
print(x % y)
```

```python
print( x ** y)
```

```python
print( x // y)
```

## Comparison operators

```python
print(x > y)
```

```python
print(x < y)
```

```python
print(x >= y)
```

```python
print(x <= y)
```

```python
print(x == y)
```

```python
print(x != y)
```

## Logical operators

```python
x = True
y = False

print( x or y)
```

```python
print(x and y)
```

```python
print(not x)
print(not y)
```

## Bitwise operators

```python
x = 2
y = 3
```

```python
print(x & y)
```

```python
print(x | y)
```

```python
print(x >> 2)
```

```python
print(y << 3)
```

```python
print(~ x)
```

## Assignment operators

```python
a = 3
print(a)
```

```python
a += 3
# a = a + 3
print(a)
```

```python
a -= 3
# a = a - 3
print(a)
```

```python
a *= 3
# a = a * 3
print(a)
```

```python
a /= 3
# a = a / 3
print(a)
```

```python
a++
++a
```

```python
a = a + 3
a += 3
```

## Identity operators

```python
a = 3
b = 3

print(a is b)
```

```python
a = "Hello"
b = "Hello"

print(a is b)
```

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a is b)
```

```python
a = "Hello-World"
b = "Hello-World"

print(a is b)
print(a is not b)
```

## Membership operators

```python
x = "Delhi"

print("D" in x)
print("D" not in x)
```

```python
x = [1, 2, 3]
 
print(1 in x)
print(5 in x)
```

```python
x = (1, 2, 3)

print(1 in x)
print(5 in x)
```
