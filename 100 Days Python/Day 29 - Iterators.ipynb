## What is an Iteration

Iteration is a general term for taking each item of something, one after another. Any time you use a loop, explicit or implicit, to go over a group of items, that is iteration.

for eg loops iterates over a sequence of data, and performs operations

```python
num = [1,2,3]

for i in num:
    print(i)
```

## What is Iterator

An Iterator is an object that allows the programmer to traverse through a sequence of data without having to store the entire data in the memory

so iterator is the one because of which iteration is possible, so when we pass
a loop then internally there's an iterator object that fetches items one by one 
and gives back to the loop.

now the biggest feature of iterator is that for it to traverse over this sequence of data  doesnt need to load the whole data into the memory, it rather works very smartly. its fetches one data at a time, loads it into the memory, performs whatever operation it needs to perform, then removes it. and then keep on repeating this whole cycle again and again. so by this it keeps only one thing in the memory at a time, due to this memory gets saved.

```python
L = [x for x in range(1,100)]

for i in L:
    print(i*2, end = ',')
```

```python
import sys

print(sys.getsizeof(L)/1024)
```

```python
x = range(1,100)

for i in x:
    print(i*2, end = ',')
```

```python
print(sys.getsizeof(x)/1024)
```

so when we use iterator to traverse a sequence of data, the iterator doesn't store the entire sequence of data into the memory, it only does upto a certain degree thats suits best for its requirement, which is one iter on which it performs the operation and then it will remove it and will load the other.
so this is a very efficient way of traversing/looping to data.

## Making our own for loop

```python
def mera_khudka_for_loop(iterable):
    
    iterator = iter(iterable)
    
    while True:
        
        try:
            print(next(iterator))
        except StopIteration:
            break  
```

```python
a = [1,2,3]
b = range(1,11)
c = (1,2,3)
d = {1,2,3}
e = {0:1,1:1}

mera_khudka_for_loop(e)
```

## Let's create our own range() function

```python
class mera_range:
    
    def __init__(self, start, end):
        self.start = start
        self.end = end
        
    def __iter__(self):
        return mera_range_iterator(self)
```

```python
class mera_range_iterator:
    
    def __init__(self, iterable_obj):
        self.iterable = iterable_obj
    
    def __iter__(self):
        return self
    
    def __next__(self):
        
        if self.iterable.start >= self.iterable.end:
            raise StopIteration
            
        current = self.iterable.start
        self.iterable.start += 1
        return current
```

```python
for i in mera_range(1, 11):
    print(i)
```

```python
x = mera_range(1,11)
```

```python
iter(x)
```

## Application of Iterators

iterator is a technique by using which we can perform an operation one byone over a stream of data without loading the entire data in the memory. its the biggest feature of iterator.

ex - working on deep learning, over a large no of image data with a huge size, now we will convert them all into numpy arrays, and then to perform some operations over it. now the problems come if we are dealing with limited ram in our system. now the problem arise to how to load the entire data into our memory and then how can we extract the numpy arrays from it. its actually not possible.

but if we use an iterator then this job is achievable.

iterator will pick that stream of data, one by one load it into the memory, extract its numpy array, and will return its next image.

so whenever if gets stuck in ML or DL over some large data, by using iterators, can create such data types that can handle a large amount of data because iterator at one point keeps only one item into the memory, and thats its biggest benefit.

also in keras library, are image data generators that exactly works like iterators and depicts the exact feature and implementation.
