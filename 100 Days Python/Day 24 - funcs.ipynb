# What are Functions?

```python
# A block of code that performs a specific task.

# functions are basically black boxes in which what code is written doesn't matter, what does matter the input it wants and the output it generates.

# functions make programming easier since they allow us to reuse code instead of rewriting it.
```

### two important things regarding functions:

### 1. Abstraction

```python
# Abstraction is used to hide the internal functionality of the function from the users.

# The users only interact with the basic implementation of the function, but inner working is hidden.

# User is familiar with that "what function does" but they don't know "how it does."
```

### 2. Decomposition

```python
# Decomposition is a way to break down your systems into modules in such a way that each module provides different functionality, but may affect other 
# modules also.
```

### Components of a Function

### Parameters Vs Arguments

```python
# Parameters:
# A parameter is the variable defined within the parentheses during function definition.
# Simply they are written when we declare a function.

# Arguments:
# An argument is a value that is passed to a function when it is called.
# It might be a variable, value or object passed to a function or method as input.
# They are written when we are calling the function.
```

1. Default Argument

3. Keyword Argument

### Global Var and Local Var

###### Examples:

```python
# functions as arguments

def func_a():
    print("inside func_a: ")
    
    # No return value so return value becomes None

def func_b(y):
    print("inside func_b: ")
    return y

def func_c(z):
    print("inside func_c: ")
    return z()

print(func_a())
print(5 + func_b(2))
print(func_c(func_a))
```

```python
# shows accessing variables output scope

def f(y):
    x = 1
    x += 1
    print(x)
    
x = 5
f(x)
print(x)

# function itself works as an independent program.
# so variables inside it and variables inside Global frame can coexist with same name.
```

```python
# variables which are defined inside a function is called Local Variable.

# variables which are not defined inside any function but are present inside main program are called Global Variable.
```

```python
def g(y):
    print(x) # printing varaible without creating, also if function doesn't have its own local variable then it can use global variable.
    print(x + 1) # it does not mean that changes were made in x, it means since its immutable, a new integer (6) formed and got printed.
                # no change in value of x in the main program
x = 5
g(x)
print(x)
```

```python
def h(y):
    x += 1 # leads to an error without line "global x" inside h
    
x = 5
h(x)
print(x)

# Rule: we can use global variable, but we cannot make any changes to it. this is not allowed.

# Concept 1: concept of global variable: a variable that's not present inside any function,
                                       # but present inside the main program that can be used by any function in the program.

# Concept 2: if a function doesn't have its own local variable(a variable that's created inside a function, 
                                       # and later gets destroyed within the function after usage), then it can use a global variable.

# Concept 3: local variable can only use global variable, but cannot make any changes to it.
```

```python
# But we can EXPLICITLY change global variable from local variable

def h(y):
    global x # Note: by using global keyword we can change global variables, but we should never do it.
    x += 1
    
x = 5
h(x)
print(x)
```

```python
# complicated scope

def f(x):
    x = x+ 1
    print("in f(x): x =", x)
    return x

x = 3
z = f(x)
print("in main proram scope: z =", z)
print("in main program scope: x =", x)
```

### Nested Functions

```python
def f():
    print("Inside f")
    def g():
        print("Inside g")
    g()
```

```python
f()
```

```python
g()

# Nested Function stays abstracted/hidden from the main program.
```

```python
def f():
    print("Inside f")
    def g():
        print("Inside g")
        f()
    g()
```

```python
f()

# Infinite Loop - Code will Crash - Kernel Dead
```

```python
# harder scope

def g(x):
    def h():
        x = "abc"
    x = x + 1
    print("in g(x): x =", x)
    h()
    return x

x = 3
z = g(x)
```

```python
# complicated scope

def g(x):
    def h(x):
        x = x + 1
        print("in h(x): x =", x)
    x = x + 1
    print("in g(x): x =", x)
    h(x)
    return x

x = 3
z = g(x)
print("in main proram scope: x =", x)
print("in main program scope: z =", z)
```
