```python
class Atm:

    __counter = 1 # static / class variable
    
    def __init__(self): # init --> Constructor
                        # Constructor is a special method(func defined inside a class),
                        # It's used to initialize the instance members of the class.
                
                        # Code written inside constructor executes automatically
                        # when object of its class is created.
        self.__pin = ""
        self.__balance = 0
        
        self.sno = Atm.counter # instance variable
        Atm.__counter = Atm.__counter + 1
        
        print(id(self))
        
        #self.__menu()
    
    @staticmethod
    def get_counter():
        return Atm.__counter
    
    @staticmethod
    def set_counter(new):
        if type(new) == int:
            Atm.__counter = new
        else:
            print('Not Allowed')
    
    def get_pin(self):
        return self.__pin
    
    def set_pin(self, new_pin):
        if type(new_pin) == str:
            self.__pin = new_pin
            print("Pin changed")
        else:
            print("Not allowed")
        
    def __menu(self):
        user_input = input("""
                    Hello, how would you like to proceed?
                    1. Enter 1 to create pin
                    2. Enter 2 to deposit
                    3. Enter 3 to withdraw
                    4. Enter 4 to check balance
                    5. Enter 5 to exit
        """)
        if user_input == "1":
            self.create_pin()
        elif user_input == "2":
            self.deposit()
        elif user_input == "3":
            self.withdraw()
        elif user_input == "4":
            self.check_balance()
        else:
            print("bye")
            
    def create_pin(self):
        self.__pin = input("Enter your pin: ")
        print("Pin set successfully")
        
    def deposit(self):
        temp = input("Enter your pin: ")
        if temp == self.__pin:
            amount = int(input("Enter the amount: "))
            self.__balance = self.__balance + amount
            print("Deposit successful")
        else:
            print("Invalid pin")
    
    def withdraw(self):
        
        temp = input("Enter your pin: ")
        if temp == self.__pin:
            amount = int(input("Enter the amount: "))
            if amount <= self.__balance:
                self.__balance = self.__balance - amount
                print("Operation successful")
            else:
                print("insufficient funds")
        else:
            print("invalid pin")
    
    def check_balance(self):
        
        temp = input("Enter your pin: ")
        if temp == self.__pin:
            print(self.__balance)
        else:
            print("invalid pin")
```

### Creating own data type (Fraction) & corresponding methods to add, sub, mul and divide fractions

```python
class Fraction:
    
    def __init__(self, n, d):
        self.num = n
        self.den = d
        
    def __str__(self):
        return"{}/{}".format(self.num,self.den)
    
    def __add__(self,other):
        
        temp_num = self.num * other.den + other.num * self.den
        temp_den = self.den * other.den
        
        return"{}/{}".format(temp_num,temp_den)
    
    def __sub__(self,other):
        
        temp_num = self.num * other.den - other.num * self.den
        temp_den = self.den * other.den
        
        return"{}/{}".format(temp_num,temp_den)
    
    def __mul__(self,other):
        
        temp_num = self.num * other.num
        temp_den = self.den * other.den
        
        return"{}/{}".format(temp_num,temp_den)
    
    def __truediv__(self,other):
        
        temp_num = self.num * other.den
        temp_den = self.den * other.num
        
        return"{}/{}".format(temp_num,temp_den)
```

```python
# benefits of Inheritence:
# In coding theres a principle "DRY" dont repeat yourself,
# thats the defination of writing good code.
# one thats already written, need not to be done again.
# thats where Inheritence comes into picture
```

```python
# Note: Inheritence always acts towards upward direction

# ex - Student can inherit from user like - login, registraion
# but user can't revert back and try to inherit from the student

# i.e child and inherit properties and data members
```

```python
# In inheritence child inherits:
#     Data Members / Variables
#     Member functions / Methods
#     Constructor

# Note: Private members are not inherited   
```

```python
class User: # parent
    
    def login(self):
        print('login')
    
    def register(self):
        print('Register')

class Student(User): # child
    
    def enroll(self):
        print('Enroll')
    
    def review(self):
        print('Review')  

stu1 = Student()

stu1.enroll()
stu1.review()
stu1.login()
stu1.register()
```

```python
# the reverse is not true
# i.e parent can't access child

class User: # parent
    
    def login(self):
        print('login')
    
    def register(self):
        print('Register')

class Student(User): # child
    
    def enroll(self):
        print('Enroll')
    
    def review(self):
        print('Review')

u = User()

u.enroll()
u.review()
u.login()
u.register()
```

# Polymorphism

```python
# Ex 3 - Polymorphism
```

```python
# Method Overriding

class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print("Buying a smartphone")

s=SmartPhone(20000, "Apple", 13)

s.buy()
```

```python
# parent class has a method, and with the same name child class has a method,
# when child class method will call that method, then method that will get
# executed will be of child class.

# if parent has the same thing thats present in child class, and the two
# are conflicting with each other, then child one will get executed.

# this is known as method overriding
```

```python
# Polymorphism:
#     Method Overriding
#     Method Overloading
#     Operator Overloading
```

```python
# Ex - class Parent
```

```python
class Parent:

    def __init__(self, num):
        self.__num = num

    def get_num(self):
        return self.__num

class Child(Parent):
    
    def show(self):
        print("This is in child class")
        
son = Child(100,)
print(son.get_num())
son.show()
```

```python
# Ex - 2
```

```python
class Parent:

    def __init__(self, num):
        self.__num = num

    def get_num(self):
        return self.__num

class Child(Parent):

    def __init__(self, val, num):
        self.__val = val

    def get_val(self):
        return self.__val
        
son = Child(100, 10)
print("Parent: Num:", son.get_num())
print("Child: Val:", son.get_val())
```

```python
# if child doesn't have its own constructer, then parent constructer get
# invoked automatically, but if child has it own constructer, then in this case
# parent's constructer will not get called
```

### User of super()

```python
# Ex - Super
```

```python
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    def buy(self):
        print ("Buying a smartphone")

        super().buy() # syntax to call parent's buy method

s=SmartPhone(20000, "Apple", 13)

s.buy()
```

```python
# super keyword doesn't work outside the class

s.super().buy()
```

```python
# super keyword can access only 2 things:
#     methods of paret class
#     constructor of parent
    
# attributes can't be accessed by super keyword
```

```python
# Ex - Super example with constructor
```

```python
class Phone:
    
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

class SmartPhone(Phone):
    
    def __init__(self, price, brand, camera, os, ram):
        print('Inside smartphone constructor')
        super().__init__(price, brand, camera)
        self.os = os
        self.ram = ram
        print ("Inside smartphone constructor")

s = SmartPhone(20000, "Samsung", 12, "Android", 2)

print(s.os)
print(s.brand)
```

```python
# here we are deregating things,
# things need to be done by child is done by child itself
# but for things the code is already written in parent class will
# not be done by child class, for those we will call the constructor to get
# done with the job

# half initialisation is done by child class, and for the rest initialisation
# we are calling the parent class using the super keyword
```

```python
# Ex - super 
```

```python
class Parent:

    def __init__(self, num):
      self.__num = num

    def get_num(self):
      return self.__num

class Child(Parent):
  
    def __init__(self, num, val):
      super().__init__(num)
      self.__val = val

    def get_val(self):
      return self.__val
      
son=Child(100, 200)
print(son.get_num())
print(son.get_val())
```

```python
class Parent:
    
    def __init__(self):
        self.num = 100

class Child(Parent):

    def __init__(self):
        super().__init__()
        self.var = 200
        
    def show(self):
        print(self.num)
        print(self.var)

son = Child()
son.show()
```

```python
# we can call parent's attribute from inside the child class using self
# this is possible because self is nothing but this object--> son
# so if this obejct can call from outside then calling from inside is also possible
```

```python
class Parent:
    
    def __init__(self):
        self.__num = 100

    def show(self):
        print("Parent:",self.__num)

class Child(Parent):
    def __init__(self):
        super().__init__()
        self.__var = 10

    def show(self):
        print("Child:", self.__var)

dad = Parent()
dad.show()
son = Child()
son.show()
```

Inheritance in summary

- A class can inherit from another class.

- Inheritance improves code reuse

- Constructor, attributes, methods get inherited to the child class

- The parent has no access to the child class

- Private properties of parent are not accessible directly in child class

- Child class can override the attributes or methods. This is called method overriding

- super() is an inbuilt function which is used to invoke the parent class methods and constructor

### MRO - Method Resolution Order

```python
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class Product:
    def buy(self):
        print ("Product buy method")

# Method Resolution Order
class SmartPhone(Product, Phone):
    pass

s=SmartPhone(20000, "Apple", 12)

s.buy()
```

```python
# The order we are inheriting, will decide whose priority is more.
# this concept is known as MRO
# In case of conflict, method of only those class will execute that we have
# inherited first
```

### Method overloading and Operator overloading

```python
# Polymorphism
# 1. Method Overriding
# 2. Method Overloading
# 3. Operator Overloading
```

```python
# "Method Overloading" means when your single method, is implemented in different
# ways, means giving different different inputs to the single method to induce
# different different behaviour
```

```python
# eg - for the code of calculating area, but different shapes has different 
# ways to calculate their area, but to make sure the good readibility if the code,
# for all the shapes whose area we want to calculate we keep their function name
# area, so that it makes clear that the function calculates area.
# we just will give different inputs, and get different behaviours.
```

```python
class Geometry:
    
    def area(self,radius):
        return 3.24 * radius * radius

    def area(self, l, b):
        return l * b
    
obj = Geometry()
print(obj.area(4))
```

```python
# this would have worked perfectly in java but in python it throwed an error,
# its because in python it doesn't work that we make multiple methods
# with the same name, to get different behaviours by giving different inputs.

# in python method overloading, techincally speaking doesn't work,
# but may work if --->
```

```python
class Geometry:
    
    def area(self,a, b = 0):
        if b == 0:
            print('Circle', 3.14 * a * a)
        else:
            print('Rectangle', a * b)
    
obj = Geometry()
obj.area(4)
obj.area(4, 5)
```

```python
# to be honest, we don't need the java feature, we can use some kind of hack
# by adding default arguments can anyways achieve those behaviour, thats the
# philosophy

# technically method overloading doesn't exit in python, but by using this kind of
# makeshifting we are able to do so.
```

### Bank Example Hierarchy 

```python
# lets take a eg - suppose we are working in a bank,
# we buit a "bank" application that acts as a top level application thats connected
# to databases and all query related stuff, below it run two more applications,
# "bank web app", and "back mobile app".

# Bank App --> Database

# Web App --> Bank App <-- Mobile App (Inheritence)

# so we sould want that the child classes (Web App and Mobile App) for
# the permission for database related things, we need to code a security 
# feature for the app, i.e we need to force a constraint on the child classes
# that to inherit the parent class, u need to write the security function also,
# because web and mobile app should be secure.

# so the whole concept of abstraction is that we as a sr. dev
# can apply constraints to our lower classes
```

### Abstract Class

```python
# Abstract Class - a class within which there is atleast one abstract method 
# is present.

# Abstract Method - a method within which there is no code present, i.e
# inside it there lies no implementation

# there are two types of methods - "Abstract Methods" and "Concrete Methods"

# Concrete Methods are those within which there is some code written and no
# code is written inside Abstract Methods
```

### Coding Abstract class Eg - BankApp  Class

```python
from abc import ABC, abstractmethod

class BankApp(ABC):

  def database(self):
    print('connected to database')

  @abstractmethod # decorator
  def security(self):
    pass
```

```python
# two thing needed for making an abstract class:
# 1. Inheriting ABC class
# 2. atleast one abstract methods should be present inside your class

# Abstract Method is a method inside which there is no code written,
# and there's a decorator is present on its top named as @abstractmethod
```

```python
class MobileApp(BankApp):

  def mobile_login(self):
    print('login into mobile')
```

```python
mob = MobileApp()
```

```python
# we can't inherit bank app till we don't add bank app's abstract method
# into our class
```

```python
class MobileApp(BankApp):
    
    def mobile_login(self):
        print('login into mobile')
    
    def security (self):
        print('mobile security')
```

```python
mob = MobileApp()
```

```python
# so we learned that by using abstraction, higher level class can comple the 
# lower level class to perform certain things, and its good because in a proper
# big application we would want that if we are creating the top level
# classes, then the lower classes can follow certain instruction
```

```python
# so in abstraction we hide the details, and its actual use case is that
# we as a sr. dev or a top level class designer we make sure that the child
# classes listen to your instructions
```

```python
# in general before giving our things we are applying a clause, and that 
# clause must be agreed by the other classes, and the clause is our
# abstract method.
```

```python
from abc import ABC,abstractmethod

class BankApp(ABC):

  def database(self):
    print('connected to database')

  @abstractmethod
  def security(self):
    pass

  @abstractmethod
  def display(self):
    pass
```
