```python
# Title Case a string without using title function

sample = "how are you?"

sample.split()

L = []

for i in sample.split():
    L.append(i.capitalize())
    
print(L)
print(" ".join(L))
```

```python
L1 = [1,1,2,2,3,3,4,4]
# Output: L2 = [1,2,3,4]

L = []

for i in L1:
    if i not in L:
        L.append(i)
    
print(L)    
```
