# Python supports 3 categories of data types:

1. Basic types: (Integer, Float, Complex, Boolean, and String)
2. Container types: (List, Tuples, Sets, and Dictionary)
3. User-defined types: Class

###### integer

```python
print(4)
print(1e308)
print(1e309)
```

###### float

```python
print(4.5)
print(1.8e307)
print(1.8e308)
print(1.9e308)
```

###### boolean

```python
print(True)
print(False)
```

###### compelx

```python
print(4+5j)
```

###### string

```python
print('Kolkata')
print("Kolkata")
print('''Kolkata''')
print("""Kolkata""")
```

###### list

```python
print([1,2,3,4,5])
```

###### tuple

```python
print((1,2,3,4,5))
```

###### sets

```python
print({1,2,3,4,5})
```

###### dict

```python
print({"Name":"Saurabh","Age":22,"Gender":"Male"})
```
