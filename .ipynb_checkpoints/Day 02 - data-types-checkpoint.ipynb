# Python supports 3 categories of data types:

1. Basic types: (Integer, Float, Complex, Boolean, and String)
2. Container types: (List, Tuples, Sets, and Dictionary)
3. User-defined types: Class

```python
# integer
print(4)
print(1e308)
print(1e309)
```

```python
# float
print(4.5)
print(1.8e307)
print(1.8e308)
print(1.9e308)
```

```python
# boolean
print(True)
print(False)
```

```python
# compelx
print(4+5j)
```

```python
# string
print('Kolkata')
print("Kolkata")
print('''Kolkata''')
print("""Kolkata""")
```

```python
# list
print([1,2,3,4,5])
```

```python
# tuple
print((1,2,3,4,5))
```

```python
# sets
print({1,2,3,4,5})
```

```python
# dict
print({"Name":"Saurabh","Age":22,"Gender":"Male"})
```
