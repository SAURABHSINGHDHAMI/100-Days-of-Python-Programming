Write a program to calculate the **sum** of the following series up to the **nth** term:

The value of **n** will be provided by the user.

```python
num = int(input('Enter your number: '))

result = 0
fact = 1

for i in range(1, num + 1):
    fact = fact * i
    result = result + (i / fact)

print(result)
```
