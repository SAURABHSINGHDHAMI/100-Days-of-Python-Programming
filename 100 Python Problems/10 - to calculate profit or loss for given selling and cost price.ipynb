```python
# Write a program that will take user input of cost price and selling price and determines whether its a loss or a profit
```

```python
cp = float(input("Enter your cost price: "))
sp = float(input("Enter your selling price: "))

if cp>sp:
    amount = cp - sp
    print("Loss: ", amount)
else:
    amount = sp - cp
    print("Profit: ", amount)
```
