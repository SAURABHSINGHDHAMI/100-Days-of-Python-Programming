Write a program to swap the key value pair for max and min values

Eg if the dict is like this {‘a’:1,’b’:2,’c’:3}

Output should be {a:3,b:2,c:1}

```python
D = {'a':100, 'b':200, 'c':300, 'd':400}
max_val = max(D.values())
min_val = min(D.values())

for i in D:
    if D[i] == max_val:
        print(D[i])
        break
    
for j in D:
    if D[j] == min_val:
        print(D[j])
        break

D[i] = min_val
D[j] = max_val
print(D)
```
