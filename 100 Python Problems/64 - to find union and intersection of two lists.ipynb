Write a program that can perform union and intersection on 2 given list.

```python
L1 = [1, 2, 3, 4, 5]
L2 = [3, 4, 5, 6, 7]
union = []
intersection = []

#union
for i in L1:
    if i not in union:
        union.append(i)
for j in L2:
    if j not in union:
        union.append(j)

# intersetion
for i in L1:
    if i in L2:
        intersection.append(i)

print('Your union list:', union)
print('Your interssection list:', intersection)
```
